## Sentiment Analysis on Titles and Tags


```python
from textblob import TextBlob

# Title sentiment
df['title_sentiment'] = df['title'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Tags sentiment
df['tags_sentiment'] = df['tags'].apply(lambda x: TextBlob(str(x).replace("|", " ")).sentiment.polarity)

df[['title', 'title_sentiment', 'tags_sentiment']].head()
```

## Creating Graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Region-wise comparison

```python
region_views = df.groupby('publish_country')['views'].mean()
plt.figure(figsize=(10,4))
sns.barplot(x=region_views.index, y=region_views.values)
plt.title('Average Views by Country', fontweight='bold')
plt.ylabel('Average Views')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

```python
top_category= df.groupby('category_id')['views'].mean().sort_values(ascending=False)
sns.barplot(x=top_category.index, y=top_category.values, color='g')
plt.ylabel('Average view')
plt.xlabel('Category_id')
plt.title('Average view by category', fontweight='bold')
plt.show()
```

## Pairplot

```python
numeric_df = df[['views', 'likes', 'dislikes', 'comment_count']]
sns.pairplot(numeric_df)
sns.set(style='whitegrid')
plt.show()
```

## Pie-Chart

```python
plt.figure(figsize=(6,6))
country_counts = df['publish_country'].value_counts()
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90)
plt.title(' Countries by Video Count', fontweight='bold')
plt.show()
```

## Line plot

```python
trend_count = df.groupby('trending_date').size()
sns.lineplot(x= trend_count.index, y= trend_count.values, color='green')
plt.xlabel('Dates')
plt.ylabel('Video_count')
plt.title('Trending videos over time', fontweight='bold')
plt.show()
```

## Histogram plot''

```python
sns.histplot(df['likes'], bins=50, kde=True, color='magenta')
plt.title('Histogram of views')
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.show()
```

## Heatmap

```python
numeric_df =df[['views', 'likes', 'dislikes', 'comment_count']]
sns.heatmap(numeric_df.corr(), annot=True,)
plt.title('Heatmap of features correlation')
plt.show()
```
